```python
import numpy as np
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv('./../lab_10/iris.csv')

X = iris[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y = iris['variety']

n=3
estimator =  KMeans(n_clusters=n).fit(X)
# estimator.fit(X)
iris['cluster'] = estimator.labels_
print(iris.head)


centers = pd.DataFrame(estimator.cluster_centers_, columns=['sepal.length', 'sepal.width', 'petal.length', 'petal.width'])
print(centers)

fig, ax = plt.subplots(figsize=(4, 4))
ax = sns.scatterplot(x='sepal.length', y='petal.length', hue='variety', style='cluster', 
                     ax=ax, data=iris)

centers.plot.scatter(x='sepal.length', y='petal.length', ax=ax, s=50, color='black')

plt.tight_layout()
plt.show()
```

```python
K=range(1,10)

KM = [KMeans(k).fit(X) for k in K]
inertia = [v.inertia_ for v in KM]   # cluster centroids



plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within-cluster sum of squares')
plt.title('Elbow for KMeans clustering')
# elbow curve
plt.plot(K, inertia, 'b*-')
```

```python
from sklearn import manifold, datasets
n_components = 2
perplexity = 5
tsne = manifold.TSNE(n_components=n_components, init='random',
                         random_state=0, perplexity=perplexity)
Y_tsne = tsne.fit_transform(X)

# ax = plt.figure()
plt.title("Perplexity=%d" % perplexity)

species = np.unique(y)
colors = 'rgbk'
for i in range(0,len(species)):
    plt.scatter(Y_tsne[y==species[i], 0], Y_tsne[y==species[i], 1], c=colors[i])

plt.axis('tight')
plt.plot()
```
